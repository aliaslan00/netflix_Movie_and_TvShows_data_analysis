```python
sns.countplot(df["type"])
fig = plt.gcf()
fig.set_size_inches(5,5)
plt.title("tür")
```

```python
sns.countplot(df["rating"])
sns.countplot(df["rating"]).set_xticklabels(sns.countplot(df["rating"]).get_xticklabels() , rotation = 90 , ha = "right" )
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title("rating")
```

```python
plt.figure(figsize = (10,8))
sns.countplot(x = "rating" , hue = "type" , data = df)
plt.title("tür ve reyting oranları")
plt.show()
```

```python
labels = ["Movie" , "TV shows"]
size = df["type"].value_counts()
colors = plt.cm.Wistia(np.linspace(0,1,2))
explode = [0,0.1]
plt.rcParams["figure.figsize"]=(9,9)
plt.pie(size, labels=labels, colors=colors, explode=explode, shadow=True, startangle=90)
plt.title("tür görünümü" , fontsize=25)
plt.legend()
plt.show()
```

```python
df["rating"].value_counts().plot.pie(autopct="%1.1f%%", shadow=True, figsize=(10, 8))
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
plt.figure(figsize=(25, 15))  # plt.subplots() kullanarak alt grafik oluşturuldu
wordcloud = WordCloud(
    background_color="white",
    width=1920,  # widht -> width düzeltildi
    height=1080  # height parametresi doğru şekilde belirtildi
).generate(" ".join(data.country))  # generite -> generate düzeltildi
plt.imshow(wordcloud)
plt.axis("off")  # Eksenleri kapatmak için axis("off") kullanıldı
plt.savefig("country.png")
plt.show()
```
